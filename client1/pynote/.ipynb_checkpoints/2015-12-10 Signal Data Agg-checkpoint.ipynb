```python
df['datetime'] = df.Firsttime.apply(lambda x: np.datetime64(x,'ms'))
```

```python
df['minTime'] = df['datetime'].apply(lambda x: x.replace(second = 0, microsecond  = 0)) 
```

```python
df['Midfrequency(100Hz)'] = (df['Midfrequency'] * 10).astype(int)
```

```python
minTimeBase =  np.min(df['minTime'])
```

```python
MidFBase = np.min(df['Midfrequency(100Hz)'])
```

```python
float(MidFBase)/10
```

```python
np.min(df['Midfrequency'])
```

```python
%%time
counts = np.zeros((234,280))
for index, row in df.iterrows():
    a = row
    cindex = int((row['minTime'] - minTimeBase) / np.timedelta64(1,"m"))
    rindex = int(row['Midfrequency(100Hz)'] - MidFBase)
#     print str(row) + " " + str(col)
    counts[rindex][cindex] = counts[rindex][cindex] + 1
    
```

```python
a['Midfrequency(100Hz)']
```

```python
counts = counts.astype(int)
```

```python
counts
```

```python
fig =plt.figure(figsize=(15,15))
plt.imshow(counts,cmap=plt.cm.Blues)
plt.title("test for agg result")
```

```python
countsList = counts.tolist()
```

```python
startTime = minTimeBase.to_datetime()
startTime = np.datetime64(startTime)
cache = {}
cache['aggCount'] = countsList
cache['minTime'] = int(startTime.astype(int)/1000)
cache['maxTime'] = cache['minTime'] + len(cache['aggCount'][0]) * 1*60*1000
cache['minMidfre'] = float(MidFBase)/10
cache['maxMidfre'] = 958.7
```

```python
cache['maxMidfre']
```

```python
cache.keys()
```

```python
f = open('../data/preCompute.json','wb')
```

```python
result = json.dumps(cache)
json.dump(result,f)
f.close()
```
